```python
#using multiple conditions
df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype(bool)
df["Gender"] = df["Gender"].astype("category")
df["Team"] = df["Team"].astype("category")
df.head()
```

```python
#filter data with more than 1 condition - AND & OR
#we want to take out female and finance employees
```

```python
var1 = df["Gender"] == "Female"
var2 = df["Team"] == "Finance"
df[var1 & var2]
```

```python
#using 'or' operator
#condition - senior managemen position or joined before 1990
var1 = df["Start Date"] < "1990-01-01"
var2 = df["Senior Management"]
df[var1 | var2]
```

```python
# condition - either name as robert & client services team or start date greater than 01-06-2016
```

```python
v1 = df["First Name"] == "Robert"
v2 = df["Team"] == "Client Services"
v3 = df["Start Date"] > "2016-06-01"
df[(v1&v2)|v3]
```

```python
#check for inclusion using isin method
#much easier than filtering using variables as done above
v1 = df["Team"].isin(["Legal", "Sales", "Product"])
df[v1]
```

```python
#.duplicated method to check for duplicate rows
df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype(bool)
df["Gender"] = df["Gender"].astype("category")
df["Team"] = df["Team"].astype("category")
df.sort_values("First Name", inplace = True)
df.head()
```
