```python
#.astype - to change dataframe column/row type and nAN values need to be dropped for this command to work, so --- ....
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba.head(3)
nba["Salary"].fillna("0", inplace = True)
nba["College"].fillna("None", inplace = True)
nba.head(8)
```

```python
nba.info()
```

```python
nba["Salary"] = nba["Salary"].astype("int")
```

```python
nba.head(4)
```

```python
#as we changed datatype from float to int, memory usgae decreases
nba.info()
```

```python
nba["Number"] = nba["Number"].astype(int)
nba["Age"] = nba["Age"].astype(int)
```

```python
#again even lower memory usage
nba.info()
```

```python
#unique object creation - category datatype in jupyter
nba["Position"].nunique()
```

```python
nba["Position"] = nba["Position"].astype("category")
```

```python
nba["Team"].nunique()
```

```python
#applying catergory data type on teams
nba["Team"] = nba["Team"].astype("category")
```

```python
#even more memory saved
nba.info()
```
