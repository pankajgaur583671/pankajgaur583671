```python
#filtering with the where method
bond = pd.read_csv("jamesbond.csv")
bond
```

```python
mask = bond["Actor"] == "Sean Connery"
bond[mask]
```

```python
bond.where(mask)
# all other datasets where sean connery isn't there gets reduced to Nan
```

```python
# another way to do it
bond.where(bond["Budget"] > 100)
```

```python
m1 = bond["Actor"] == "Sean Connery"
m2 = bond["Budget"] > 50
bond.where(m1 & m2)
```
